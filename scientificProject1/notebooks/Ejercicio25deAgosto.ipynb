```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2
import matplotlib.pyplot as plt3
import matplotlib.pyplot as plt4
import matplotlib.pyplot as plt5

def resolver_espacios(valor):
    valor = str(valor)
    if valor[0] == ' ':
        valor = valor[1:]
    return valor


def arreglar_nivel_get(valor):
    valor = str(valor)
    valor =valor.upper()
    return valor

csv=pd.read_csv("../data/Edificios.csv")
csv.dropna(inplace=True,subset=['codigo_postal'])
csv=csv.applymap(resolver_espacios)
csv['nivel_get_2']=csv['nivel_get_2'].apply(arreglar_nivel_get)
csv['nivel_get_1']=csv['nivel_get_1'].apply(arreglar_nivel_get)
```

```python
x_barrios=csv['barrio'].value_counts().index.tolist()
y_barrios=csv['barrio'].value_counts().tolist()

plt1.figure(figsize=(20,12))
plt1.xticks(rotation=90,fontsize=30)
plt1.yticks(fontsize=30)
plt1.bar(x_barrios,y_barrios)
plt.figtext(.5,.9,'Cantidad de edificios por barrios', fontsize=80, ha='center')

plt1.show()
```

```python
x_calle=csv['calle'].value_counts().index.tolist()
y_calle=csv['calle'].value_counts().tolist()


plt2.figure(figsize=(100,30))
plt2.bar(x_calle,y_calle)
plt2.xticks(rotation=90)
plt2.yticks(fontsize=30)
plt2.figtext(.5,.9,'Cantidad de edificios por calle', fontsize=200, ha='center')
plt2.tick_params(axis='x', which='major', labelsize=70)

plt2.show()
```
