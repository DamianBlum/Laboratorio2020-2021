```python
import pandas as pd

df_barrios = pd.read_csv("../data/28octubre/barrios.csv")
df_comunas = pd.read_csv("../data/28octubre/comunas.csv")
df_casos = pd.read_csv("../data/28octubre/casos.csv")
```

```python
condition_1 = df_barrios.comuna == 3
df_comuna_3 = df_barrios[condition_1]

print("1) En la comuna 3 se encuentran los diferentes barrios")
for i in df_comuna_3["barrio"].index:
    print("El barrio ", df_comuna_3["barrio"][i], "con el perimetro",
          df_comuna_3["perimetro"][i], "y el area", df_comuna_3["area"][i])
```

```python
condition_2 = (df_barrios["comuna"] > 4) & (df_barrios["comuna"] < 10)
df_comunas_entre_4_y_10 = df_barrios[condition_2]
print("2) De la comuna 5 a la 9 se encuentran los diferentes barrios")
for i in df_comunas_entre_4_y_10["barrio"].index:
    print("El barrio ", df_comunas_entre_4_y_10["barrio"][i], "con el perimetro",
          df_comunas_entre_4_y_10["perimetro"][i], "y el area", df_comunas_entre_4_y_10["area"][i])
```

```python
condition_3 = df_comunas["ID"] == 5
df_comuna_5 = df_comunas[condition_3]
print("3) La comuna 5")
for i in df_comuna_5["ID"].index:
    print("La comuna", df_comuna_5["ID"][i], "con el perimetro",
          df_comuna_5["PERIMETRO"][i], "y el area", df_comuna_5["AREA"][i])
```

```python
condition_4 = df_barrios.barrio.str.startswith('V')
df_barrios_v = df_barrios[['comuna']][condition_4]
df_barrios_v = df_barrios_v.comuna.unique()
print("4) Las comunas con barrios que comienzan con V")
for i in df_barrios_v:
    print("Comuna", i)
```

```python
condition_5 = df_barrios["barrio"].str.contains("A")
df_barrios_v = df_barrios[condition_5]
print("5) Los barrios que contienen con A")
for i in df_barrios_v["barrio"].index:
    condition_5 = df_comunas["ID"] == df_barrios_v["comuna"][i]
    df_comunas_barrios_v = df_comunas[condition_5]
    for j in df_comunas_barrios_v["ID"].index:
        print("El barrio",df_barrios_v["barrio"][i],"pertenece a la comuna", df_comunas_barrios_v["ID"][j],"y tiene", df_comunas_barrios_v["AREA"][j],"de area")
```

```python
condition_6 = df_comunas.ID == 15
df_comuna_id_15 = df_comunas[condition_6]

for i in df_comuna_id_15["PERIMETRO"].index:
    print("6) En la comuna con id 15 se tiene el siguiente perimetro",
          df_comuna_id_15["PERIMETRO"][i])
```

```python
condition_10 = df_casos.tipo_contagio == "En Investigación"
df_casos_en_investigacion = df_casos[condition_10]
df_barrios_en_investigacion = df_barrios[df_barrios.barrio.isin(df_casos_en_investigacion.barrio)]

print("10) Los barrios que tienen casos de tipo de contagio En Investigación")
for i in df_barrios_en_investigacion["barrio"].index:
    condition_10_barrio = df_barrios.barrio == df_barrios_en_investigacion["barrio"][i]
    print("El barrio", df_barrios_en_investigacion["barrio"][i], "con el indice", i)
```
